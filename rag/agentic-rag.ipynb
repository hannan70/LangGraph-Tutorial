```python
import os
from dotenv import load_dotenv
load_dotenv()

# import necessery tools
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from langchain_groq import ChatGroq
from IPython.display import Image, Markdown
from pydantic import BaseModel, Field
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_community.document_loaders import WebBaseLoader
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
import operator
from langchain_core.tools.retriever import create_retriever_tool
```

```python
os.environ['GROQ_API_KEY'] = os.getenv("GROQ_API_KEY")
```

```python
llm = ChatGroq(model="llama-3.3-70b-versatile")
```

```python
llm.invoke("what is agentic ai")
```

```python
urls = [
    "https://blog.langchain.com/langgraph/",
    "https://blog.langchain.com/building-langgraph/"
]

loader = WebBaseLoader(urls)
docs = loader.load()
```

```python
# split document
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
documents = text_splitter.split_documents(docs)
documents
```

```python
len(documents)
```

```python
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
```

```python
langGraph_vectorstore = FAISS.from_documents(documents=documents, embedding=embeddings)
langGraph_vectorstore.save_local("langGraph_faiss_vectorstore")
```

```python
### Load the vector store db
load_langGraph_vectorstore = FAISS.load_local("langGraph_faiss_vectorstore", embeddings, allow_dangerous_deserialization=True)
```

```python
langGraph_retriever = load_langGraph_vectorstore.as_retriever(search_type="similarity", search_kwargs={"k": 3})
```

```python
langGraph_retriever.invoke("what is AI")
```

```python
# create retriever tool
langGraph_retriever_tool = create_retriever_tool(
    langGraph_retriever,
    name="langGraph_retriever_vector_db",
    description="Search and run about langGraph"
    )
```

```python
langGraph_retriever_tool
```

```python
urls = [
    "https://blog.jetbrains.com/pycharm/2026/02/langchain-tutorial-2026/",
]

loader = WebBaseLoader(urls)
docs = loader.load()
```

```python
# text splitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
documents = text_splitter.split_documents(docs)

documents
```

```python
langchain_vectorstore = FAISS.from_documents(documents=documents, embedding=embeddings)
langchain_vectorstore.save_local("langchain_vectorstore")
```

```python
langchain_retriever = loaded_langchain_vectorstore.as_retriever()
```

```python
langchain_retriever_tool = create_retriever_tool(
    langchain_retriever,
    name="Langchain_retriever_db",
    description="Search about langChain"
)
```
